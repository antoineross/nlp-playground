**Working with local and remote datasets**

- CSV : load_dataset("csv", data_files="my_file.csv")
- Txt : load_dataset("text", data_files="my_file.txt")
- JSON : load_dataset("json", data_files="my_file.jsonl")
- Pickled DataFrames : load_dataset("pandas", data_files="my_dataframe.pkl")
